```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
forest=pd.read_csv('forestfires.csv')
```

```python
forest
```

```python
forest.isnull().any()
```

```python
forest.duplicated().any()
```

```python
forest.drop_duplicates(inplace=True)
```

```python
forest
```

```python
forest.drop(forest.columns[11:29],axis=1,inplace=True)
```

```python
forest
```

```python
forest1=forest.drop(['monthsep'],axis=1)
forest1
```

```python
from sklearn.preprocessing import LabelEncoder
lab_enc=LabelEncoder()
forest1['month']=lab_enc.fit_transform(forest1['month'])
forest1['day']=lab_enc.fit_transform(forest1['day'])
forest1['size_category']=lab_enc.fit_transform(forest1['size_category'])
```

```python
forest1
```

```python
target=forest1[['size_category']]
target
```

```python
features=forest1.drop('size_category',axis=1)
```

```python
features
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler()
features=std_scaler.fit_transform(features)
features
```

```python
x_train,x_test,y_train,y_test= train_test_split(features,target,train_size=0.75,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
svc= SVC(C=1.0, kernel='poly',gamma=0.1,degree=1)
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred= svc.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_pred,y_test)
```

```python
params={'C':[1,2,3,4,56,10,20],'kernel':['rbf','poly','linear','sigmoid'],'gamma':[0.1,0.2,0.3,0.4],'degree':[1,2,3,4,5]}
```

```python
grid_search= GridSearchCV(svc,params,verbose=2)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
svc= SVC(C=56, kernel='linear',gamma=0.1,degree=1)
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_pred,y_test)
```

```python
conf=confusion_matrix(y_test,y_pred)
conf
```

```python
print(classification_report(y_pred,y_test))
```
