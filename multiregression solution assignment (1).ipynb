```python
import pandas as pd
import numpy as mp
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
df= pd.read_csv('ToyotaCorolla.csv',encoding='latin1')
df
```

```python
df=df.rename({'Age_08_04':'age','Doors':'doors','Price':'price','Weight':'weight','Gears':'gears','Quarterly_Tax':'quaterlytax'},axis=1)
df
```

```python
df.isnull().sum()
```

```python
data=pd.concat([df.iloc[:,2:4],df.iloc[:,6:7],df.iloc[:,8:9],df.iloc[:,12:14],df.iloc[:,15:18]],axis=1)
data
```

```python
data.duplicated().any()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
target=data[['price']]
target
```

```python
features=data.drop('price',axis=1)
features
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.7,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lin_model1=LinearRegression()
```

```python
lin_model1.fit(x_train,y_train)
```

```python
y_pred=lin_model1.predict(x_test)
y_pred
```

```python
lin_model1.intercept_
```

```python
lin_model1.coef_
```

```python
target=df1[['Profit']]
```

```python
target
```

```python
features=df1.drop('Profit',axis=1)
```

```python
features
```

```python
features=pd.get_dummies(features)
features
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.7,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lin_model=LinearRegression()
```

```python
lin_model.fit(x_train,y_train)
```

```python
y_pred=lin_model.predict(x_test)
```

```python
y_pred
```

```python
lin_model.intercept_
```

```python
lin_model.coef_
```

```python
lin_model.score(x_test,y_test)
```

```python
sse=((y_test-y_pred)**2).sum()
sse
```

```python
sst=((y_test-y_test.mean())**2).sum()
sst
```

```python
Rsquared= 1-(sse/sst)
Rsquared
```
