```python
import pandas as pd
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
df=pd.read_csv('book (1).csv')
```

```python
df
```

```python
frequent=apriori(df,min_support=0.1,use_colnames=True)
frequent
```

```python
result=association_rules(frequent,metric='confidence',min_threshold=0.6)
result
```

```python
result.sort_values('lift',ascending=False)
```

```python
lift=result[result.lift>1]
lift
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.scatter(result['support'], result['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

```python
lift[['support','confidence']].corr()
sns.heatmap(lift[['support','confidence']].corr(),annot= True)
plt.show()
```

```python
matrix = lift.pivot('antecedents','consequents','lift')
matrix
```

```python
plt.figure(figsize=(20,6),dpi=250)
sns.heatmap(matrix,annot=True)
plt.title('HeatMap')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```
