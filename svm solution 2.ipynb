```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
df1=pd.read_csv('SalaryData_Train(1).csv')
df1
```

```python
df1.drop_duplicates(inplace=True)
```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder
df = df.apply(LabelEncoder().fit_transform)
df
```

```python
df1 = df1.apply(LabelEncoder().fit_transform)
df1
```

```python
x_train=df1.drop('Salary',axis=1)
y_train=df1[['Salary']]
x_test=df.drop('Salary',axis=1)
y_test=df[['Salary']]
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVC
svc= SVC(C=2.0, kernel='rbf',gamma=0.1)
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
model_linear = SVC(kernel = "linear")
model_linear.fit(x_train,y_train)
y_pred_linear = model_linear.predict(x_test)
accuracy_score(y_test, y_pred_linear)
```

```python
model_poly = SVC(kernel = 'poly')
model_poly.fit(x_train,y_train)
pred_test_poly = model_poly.predict(x_test)
accuracy_score(y_test, pred_test_poly)
```

```python
confusion_matrix(y_test, pred_test_poly)
```

```python
model_sigmoid = SVC(kernel = 'sigmoid')
model_sigmoid.fit(x_train,y_train)
pred_test_sigmoid = model_sigmoid.predict(x_test)
accuracy_score(y_test, pred_test_sigmoid)
```

```python
confusion_matrix(y_test, pred_test_poly)
```
